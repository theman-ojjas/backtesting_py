```python
position = False
buydates, selldates = [], []
buyprices, sellprices = [], []
Netvalue = 100000
trades = []

for index, row in data.iterrows():
    n = Netvalue / row['Close']

    if not position and row['signal'] == 1:
        buydates.append(index)
        buyprices.append(row['Close'])
        position = True

    if position and row['signal'] == 2:
        selldates.append(index)
        sellprices.append(row['Open'])
        position = False

        # Calculate the profit or loss for the trade and update Netvalue
        trade_profit = (sellprices[-1] - buyprices[-1]) * n
        Netvalue += trade_profit

        # Store the trade details in the 'trades' list
        trades.append({
            'Buy Date': buydates[-1],
            'Sell Date': selldates[-1],
            'Buy Price': buyprices[-1],
            'Sell Price': sellprices[-1],
            'Trade Profit': trade_profit,
            'Net Value': Netvalue


        })

# if position:
#     final_sell_date = data.index[-1]  # Use the last date in your dataset
#     final_sell_price = data['Close'].iloc[-1]  # Use the closing price at the last date

#     # Calculate the profit or loss for the final trade
#     final_trade_profit = (final_sell_price - buyprices[-1]) * n
#     Netvalue += final_trade_profit

#     # Store the final trade details in the 'trades' list
#     trades.append({
#         'Buy Date': buydates[-1],
#         'Sell Date': final_sell_date,
#         'Buy Price': buyprices[-1],
#         'Sell Price': final_sell_price,
#         'Trade Profit': final_trade_profit,
#         'Net Value': Netvalue
#     })


```

```python
def drawdown(list):
    acc_max = np.maximum.accumulate(list)
    return (list - acc_max).min()
```

```python
def portfolio_value(trades, field):
    """
    Calculate the portfolio value for each trade based on the specified field (e.g., 'Net Value').

    Args:
        trades (list of dict): List of trade records with 'Net Value' or a similar field.
        field (str): The field to use for calculating the portfolio value.

    Returns:
        list of float: Portfolio values at each trade.
    """
    portfolio_values = [trade[field] for trade in trades]
    return portfolio_values
```

```python
# Strategy Return
strategy_return = (Netvalue - 100000) / 100000

# Annualized Return (assuming a 60-day period)
annualized_return = (strategy_return / 60) * 365

# Number of Executed Trades
num_executed_trades = len(trades)

# Win Ratio
winning_trades = sum(1 for trade in trades if trade['Trade Profit'] > 0)
win_ratio = winning_trades / num_executed_trades

# Loss-Making Trades
loss_making_trades = sum(1 for trade in trades if trade['Trade Profit'] < 0)

# Largest Loss-Making Trade
largest_loss_trade = min(trades, key=lambda trade: trade['Trade Profit'])

# Largest Profit-Making Trade
largest_profit_trade = max(trades, key=lambda trade: trade['Trade Profit'])

# Calculate daily returns
daily_returns = [0]  # Initialize with 0 for the first day
for i in range(1, len(trades)):
    daily_returns.append((trades[i]['Net Value'] - trades[i - 1]['Net Value']) / trades[i - 1]['Net Value'])

# Benchmark Return (You'll need to fetch and calculate the benchmark data)

# Maximum Drawdown (You can use your existing 'drawdown' function)
max_drawdown = drawdown(portfolio_value(trades, 'Net Value')) / 1000


```
